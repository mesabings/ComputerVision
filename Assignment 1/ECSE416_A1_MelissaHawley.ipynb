```
#Read in the image
image = mpimg.imread('/content/drive/My Drive/ECSE416_ComputerVision/A1/images/gaussian_lighthouse.png')
image_original = mpimg.imread('/content/drive/My Drive/ECSE416_ComputerVision/A1/images/lighthouse.png')

#Convert image to grayscale for filtering
imagae_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
image_original_gray = cv2.cvtColor(image_original, cv2.COLOR_RGB2GRAY)

#filter imagee ussing 5x5 Guassian with varience of 2
image_g = cv2.GaussianBlur(imagae_gray,(5,5), 2)

#filter imagee ussing 5x5 Box with varience of 2
image_b = cv2.boxFilter(imagae_gray, -2, (5, 5))

#plot the images
plt.figure(figsize=(12,12))
plt.subplot(131),plt.imshow(image_original_gray,cmap='gray')
plt.title("Original"),plt.xticks([]),plt.yticks([])
plt.subplot(132),plt.imshow(image_g,cmap='gray')
plt.title("5x5 Guassian"),plt.xticks([]),plt.yticks([])
plt.subplot(133),plt.imshow(image_b,cmap='gray')
plt.title("5x5 Box"),plt.xticks([]),plt.yticks([])
plt.show()

#Comparing the PSNR
psnr_img_v_g = cv2.PSNR(image_original_gray, image_g)
psnr_img_v_b = cv2.PSNR(image_original_gray, image_b)

print("psnr_image_g",psnr_img_v_g)
print("psnr_image_b",psnr_img_v_b)
```

# 4.1-4. Canny Edge Detector 

*Briefly describe the 4 main steps of Canny edge detection.*
1. Filter image with x, y derivatives of Gaussian
2. Find magnitude and orientation of gradient
3. Non-maximum suppression:
    
    Thin multi-pixel wide ridges down to single pixel width
4. Thresholding and linking (hysteresis):

    Define two thresholds: low and high

```
#Read the image
checkerboard = cv2.imread('/content/drive/My Drive/ECSE416_ComputerVision/A1/images/checkerboard.png', cv2.IMREAD_GRAYSCALE)

#Step 1 blur
checkerboard_blur =  cv2.blur(checkerboard,(5,5))

#Step 2 Compute Square of derivatives
dev_x = cv2.Sobel(checkerboard_blur, cv2.CV_64F, 1, 0, ksize=3) 
dev_y = cv2.Sobel(checkerboard_blur, cv2.CV_64F, 0, 1, ksize=3) 

dev_xx = dev_x*dev_x
dev_xy = dev_x*dev_y
dev_yy = dev_y*dev_y

gdev_xx = cv2.GaussianBlur(dev_xx,(5,5),1)
gdev_xy = cv2.GaussianBlur(dev_xy,(5,5),1)
gdev_yy = cv2.GaussianBlur(dev_yy,(5,5),1)

det = (gdev_xx * gdev_yy) - (gdev_xy * gdev_xy)
trace = (gdev_xx + gdev_yy)**2
har = det - (0.05*trace)

#normalizing image pixels between zero and two 
cv2.normalize(har, har, 0, 2, cv2.NORM_MINMAX)
TRESH = 1.5
checkerboard_copy = checkerboard.copy()

#finding all the loctions where the image is gretter than the threshold
locationT = np.where(har >= TRESH)

#plotting the points on to the copied image
for pt in zip(*locationT[::-1]):
  cv2.circle(checkerboard_copy, pt,8, (0,0,255), -1)

#display images
plt.figure(figsize=(12,12))
plt.subplot(121),plt.imshow(checkerboard,cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(checkerboard_copy,cmap='gray'),plt.title('Checkerboard Harris Coorners')
plt.xticks([]), plt.yticks([])
plt.show()
```
